```python
import pgeocode
import json
from timezonefinder import TimezoneFinder

OUTPUT_FILE = "zip-2-timezone.json"
```

```python

nomi = pgeocode.Nominatim("us")
tf = TimezoneFinder()

df = nomi._data[["postal_code", "latitude", "longitude"]].dropna()

print(f"Processing {len(df)} zipcodes...")

lookup = {}
for i, (_, row) in enumerate(df.iterrows()):
    tz = tf.timezone_at(lat=row["latitude"], lng=row["longitude"])
    if tz:
        lookup[row["postal_code"]] = tz
    if i % 5000 == 0:
        print(f"  {i}/{len(df)}")

with open(OUTPUT_FILE, "w") as f:
    json.dump(lookup, f, separators=(",", ":"))

print(f"Done. {len(lookup)} entries written.")
```
